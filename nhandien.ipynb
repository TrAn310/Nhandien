```python
#Contour
#B1: Tìm canny
import cv2

path = r'f:\NCKH_2026\lapTrinhPy\Pic_Vid\cam.jpg' #r là lệnh lấy đường dẫn của ảnh cần xử lý và gán với 1 biến đặt là path 
image1 = cv2.imread(path)
image2=cv2.resize(image1, dsize= None,fx=0.1,fy=0.1)


image = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)


canny = cv2.Canny(image, 100, 220) 

#B2: Tìm contour
contours, _ = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(image2, contours, -1,(0,255,0), 1)


cv2.imshow('Image',image2)
cv2.imshow('Canny',canny)
print(' ',_ )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Bounding box
import cv2

path = r'f:\NCKH_2026\lapTrinhPy\Pic_Vid\bien_so_xe.jpg' #r là lệnh lấy đường dẫn của ảnh cần xử lý và gán với 1 biến đặt là path 
img1 = cv2.imread(path)

img=cv2.resize(img1,dsize=None,fx=0.3,fy=0.3) 



gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 100, 220, cv2.THRESH_BINARY_INV)

# Tìm contours
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    # Lấy tọa độ bounding box cho mỗi contour
    x, y, w, h = cv2.boundingRect(cnt)
    
    # Lọc bỏ các box quá nhỏ (nhiễu)
    if w > 80 and h > 80:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 1)

cv2.imshow('Result', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mini project 1/ đếm vật thể (đếm quả cam )
# Xử lý với video bằng cách tìm ngưỡng tự động (OTSU)

import cv2


path = r'F:\NCKH_2026\lapTrinhPy\Pic_Vid\cam.mp4' 
cap = cv2.VideoCapture(path)

while cap.isOpened():
    ret, frame1 = cap.read()
    if not ret:
        break

    #Tiền xử lý 
    def resize(image,scale = 0.3):
        hight = int(frame1.shape[0] * scale)
        width = int(frame1.shape[1] * scale)
        return cv2.resize(image,(width,hight))
    frame = resize (frame1)
    gray_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray_image, (5, 5), 0)
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

# khoảng màu
    lower_orange = (5, 100, 100)
    upper_orange = (20, 255, 255)

    mask = cv2.inRange(hsv, lower_orange, upper_orange)

# làm sạch mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    count = 0
    for cnt in contours:
        area = cv2.contourArea(cnt)

        if area > 1200:  # 
            x, y, w, h = cv2.boundingRect(cnt)
            count += 1
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame, f"Cam {count}", (x, y-5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)


    #Hiển thị khung hình màu gốc đã có box
    cv2.imshow('Dem so luong ', frame)

    if cv2.waitKey(20) & 0xFF == ord(' '):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Mini project 1/ Nhận diện biển số đơn giản 
#Xử lý với video có sẵn
import cv2
import numpy as np

# 1. Tiền xử lý 
path = r'F:\NCKH_2026\lapTrinhPy\Pic_Vid\xe_may.mp4'
cap = cv2.VideoCapture(path)

while cap.isOpened():
    ret, frame1 = cap.read()
    if not ret:
        break

    def resize(image,scale = 0.3):
        hight = int(frame1.shape[0] * scale)
        width = int(frame1.shape[1] * scale)
        return cv2.resize(image,(width,hight))
    frame = resize (frame1)
    gray_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray_image, (5, 5), 0)
    edges = cv2.Canny(blur, 120, 255)


# 2. Nhấn mạnh các cạnh dọc (Blackhat/Sobel)

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
    blackhat = cv2.morphologyEx(gray_image, cv2.MORPH_BLACKHAT, kernel)

# 3. Tìm cạnh bằng Sobel X (Chuyên tìm nét dọc)
    sobelx = cv2.Sobel(blackhat, cv2.CV_8U, 1, 0, ksize=3)
    sobelx = cv2.convertScaleAbs(sobelx)
    sobelx = cv2.GaussianBlur(sobelx, (5, 5), 0)

# 4. Phân ngưỡng và Đóng (Closing) để nối các nét dọc thành 1 khối
# Dùng THRESH_OTSU để máy tự tìm ngưỡng tối ưu
    ret, thresh = cv2.threshold(sobelx, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    kernel_close = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 10))
    closed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel_close)

# 5. Tìm Contours và Lọc 
    cnts, _ = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    output = frame.copy()
    count=0
    for c in cnts:
        x, y, w, h = cv2.boundingRect(c)
        aspect_ratio = w / float(h)
        area = cv2.contourArea(c)


    # Lọc tỉ lệ biển số 

        if 1800 < area < 19000:
            if (0.8 < aspect_ratio < 2.0):
                cv2.rectangle(output, (x, y), (x + w, y + h), (0, 255, 0), 2)
                count += 1
                cv2.putText(output, f"Bien so {count}", (x, y-5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)

    cv2.imshow('Bien so xe', output)
    
    # Nhấn space để thoát
    if cv2.waitKey(20) & 0xFF == ord(' '):
        break


cap.release()
cv2.destroyAllWindows()
```
